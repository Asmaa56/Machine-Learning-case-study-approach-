```python
import numpy as np 
import pandas as pd 
%matplotlib inline
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set(color_codes = True)
```

# Load house prices data

```python
df = pd.read_csv('home_data.csv')
type(df)
print(df.columns)
```

```python
data = df.drop(['id','date'], axis = 1)
```
